```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
save_model_path = './6p_4c/model_checkpoint'

A = np.load(os.path.join(save_model_path, 'train_loss.npy'))
B = np.load(os.path.join(save_model_path, 'val_loss.npy'))
M1_1 = np.load(os.path.join(save_model_path, 'train_f1.npy'))
M1_2 = np.load(os.path.join(save_model_path, 'val_f1.npy'))
M2_1 = np.load(os.path.join(save_model_path, 'train_acc.npy'))
M2_2 = np.load(os.path.join(save_model_path, 'val_acc.npy'))
M3_1 = np.load(os.path.join(save_model_path, 'train_prec.npy'))
M3_2 = np.load(os.path.join(save_model_path, 'val_prec.npy'))
M4_1 = np.load(os.path.join(save_model_path, 'train_rec.npy'))
M4_2 = np.load(os.path.join(save_model_path, 'val_rec.npy'))
epochs = len(A)

fig = plt.figure(figsize=(20, 15))
plt.subplot(321)
plt.plot(np.arange(1, epochs + 1), A)
plt.plot(np.arange(1, epochs + 1), B)
plt.title("Losses")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Train', 'Val'], loc="upper right")

plt.subplot(322)
plt.plot(np.arange(1, epochs + 1), M1_1)
plt.plot(np.arange(1, epochs + 1), M2_1)
plt.plot(np.arange(1, epochs + 1), M3_1)
plt.plot(np.arange(1, epochs + 1), M4_1)
plt.title("Train Scores")
plt.xlabel('Epochs')
plt.ylabel('Metric')
plt.legend(['Train F1', 'Train Acc', 'Train Prec', 'Train Rec'], loc="lower right")

plt.subplot(323)
plt.plot(np.arange(1, epochs + 1), M1_2)
plt.plot(np.arange(1, epochs + 1), M2_2)
plt.plot(np.arange(1, epochs + 1), M3_2)
plt.plot(np.arange(1, epochs + 1), M4_2)
plt.title("Val Scores")
plt.xlabel('Epochs')
plt.ylabel('Metric')
plt.legend(['Val F1', 'Val Acc', 'Val Prec', 'Val Rec'], loc="lower right")
plt.show()
```

```python
max_f1 = np.max(M1_2)
max_f1_epoch = np.argmax(M1_2) + 1

print(f"Max Val F1: {max_f1:.4f} at epoch {max_f1_epoch}")
```

```python
print(f"Max Score at epoch {max_f1_epoch}:")
print(f"   Train F1: {M1_1[max_f1_epoch - 1]:.4f}")
print(f"   Train Acc: {M2_1[max_f1_epoch - 1]:.4f}")
print(f"   Train Prec: {M3_1[max_f1_epoch - 1]:.4f}")
print(f"   Train Rec: {M4_1[max_f1_epoch - 1]:.4f}")
print(f"   Val F1: {M1_2[max_f1_epoch - 1]:.4f}")
print(f"   Val Acc: {M2_2[max_f1_epoch - 1]:.4f}")
print(f"   Val Prec: {M3_2[max_f1_epoch - 1]:.4f}")
print(f"   Val Rec: {M4_2[max_f1_epoch - 1]:.4f}")
```
